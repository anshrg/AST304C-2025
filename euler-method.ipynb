```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x, t: x
df = lambda x, t: -1 * x**3 + np.sin(t)

def euler_step(f, x, t, dt):
    return x + f(x, t) * dt
```

```python
t = np.linspace(0, 10, 100)
t = np.sin(np.linspace(0, np.pi/2, 100))**2 *10
x = np.zeros_like(t)
x0 = 0
```

```python
x[0] = x0
for i in range(1, len(t)):
    x[i] = euler_step(df, x[i-1], t[i-1], t[i] - t[i-1])
```

```python
plt.plot(np.linspace(0, 10, 100), t)
```

```python
# plt.plot(t, x)
plt.plot(t, x, 'o-')
```
